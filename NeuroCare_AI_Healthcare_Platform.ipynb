```python
import csv, numpy as np, os
from ultralytics.engine.results import Results

out_dir = "runs/infer/brain_tumor_test"
csv_path = os.path.join(out_dir, "detections.csv")

# model.predict above already saved images; loop through results and write CSV
rows = []
for r in pred:
    # r: ultralytics.engine.results.Results
    im_path = r.path
    if r.boxes is None:
        continue
    for b in r.boxes:
        xyxy = b.xyxy[0].cpu().numpy().tolist()
        conf = float(b.conf[0].cpu().numpy())
        cls  = int(b.cls[0].cpu().numpy())
        rows.append([im_path, *xyxy, conf, cls, r.names[cls]])

with open(csv_path, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["image","x1","y1","x2","y2","confidence","class_id","class_name"])
    w.writerows(rows)

csv_path
```

```python
import os, pickle, faiss, fitz, torch
from sentence_transformers import SentenceTransformer
from transformers import T5ForConditionalGeneration, T5Tokenizer

EMB_MODEL = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
tok = T5Tokenizer.from_pretrained("google/flan-t5-base")
llm = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")

def read_pdf_text(pdf_path):
    text = []
    doc = fitz.open(pdf_path)
    for p in doc:
        text.append(p.get_text())
    return "\n".join(text)

def chunk_text(text, chunk_size=900, overlap=150):
    chunks, start = [], 0
    while start < len(text):
        end = min(len(text), start + chunk_size)
        chunks.append(text[start:end])
        start = end - overlap
        if start < 0: start = 0
    return [c for c in chunks if c.strip()]

# Build/refresh index from uploaded PDFs (list of paths)
def build_pdf_index(pdf_paths, index_dir="/content/pdf_index"):
    os.makedirs(index_dir, exist_ok=True)
    chunks, meta = [], []
    for path in pdf_paths:
        raw = read_pdf_text(path)
        cs = chunk_text(raw)
        chunks.extend(cs)
        meta.extend([{"source": os.path.basename(path)}]*len(cs))
    embs = EMB_MODEL.encode(chunks, convert_to_numpy=True, show_progress_bar=True)
    d = embs.shape[1]
    index = faiss.IndexFlatIP(d)
    faiss.normalize_L2(embs)
    index.add(embs)

    # persist
    faiss.write_index(index, f"{index_dir}/faiss.index")
    with open(f"{index_dir}/store.pkl", "wb") as f:
        pickle.dump({"chunks": chunks, "meta": meta}, f)
    return index_dir

# Ask the index
def ask_pdf_index(question, index_dir="/content/pdf_index", k=5, max_ctx=1800):
    index = faiss.read_index(f"{index_dir}/faiss.index")
    with open(f"{index_dir}/store.pkl", "rb") as f:
        store = pickle.load(f)
    q_emb = EMB_MODEL.encode([question], convert_to_numpy=True)
    faiss.normalize_L2(q_emb)
    D, I = index.search(q_emb, k)
    ctx = "\n\n".join(store["chunks"][i] for i in I[0])

    prompt = (
      "Answer the question based ONLY on the context.\n"
      "If unsure, say you don't know.\n\n"
      f"Context:\n{ctx[:max_ctx]}\n\n"
      f"Question: {question}\nAnswer:"
    )
    inp = tok(prompt, return_tensors="pt", truncation=True, max_length=1024)
    out = llm.generate(**inp, max_new_tokens=200)
    return tok.decode(out[0], skip_special_tokens=True)
```

```python
import gradio as gr

pdf_paths_state = []

def add_pdfs(files):
    global pdf_paths_state
    pdf_paths_state = [f.name for f in files]
    build_pdf_index(pdf_paths_state)
    return f"Indexed {len(pdf_paths_state)} PDF(s)."

def ask(q):
    if not pdf_paths_state:
        return "Please upload and index PDFs first."
    return ask_pdf_index(q)

with gr.Blocks() as pdf_chat:
    gr.Markdown("## üìë Medical Report Analyzer (RAG)")
    with gr.Row():
        up = gr.Files(label="Upload medical reports (PDF)", file_types=[".pdf"])
        btn = gr.Button("Build / Refresh Index")
    status = gr.Markdown()
    q = gr.Textbox(label="Ask a question about the reports")
    ans = gr.Markdown()
    btn.click(add_pdfs, up, status)
    q.submit(ask, q, ans)

pdf_chat.launch()
```

```python
!pip install -q pandas
from ultralytics import YOLO
import pandas as pd
from PIL import Image

MED_MODEL_PATH = "/content/drive/MyDrive/medicine_model.pt"   # put your trained model here
DRUG_DB_CSV   = "/content/drive/MyDrive/drug_info.csv"        # columns: name,dosage,side_effects

drug_db = pd.read_csv(DRUG_DB_CSV) if os.path.exists(DRUG_DB_CSV) else pd.DataFrame(columns=["name","dosage","side_effects"])
med_model = YOLO(MED_MODEL_PATH) if os.path.exists(MED_MODEL_PATH) else None

def med_detect_and_lookup(img):
    if med_model is None:
        return None, "‚ö†Ô∏è medicine_model.pt not found. Train or place the file, then retry."
    res = med_model.predict(img, conf=0.5)
    ann = Image.fromarray(res[0].plot()[..., ::-1])  # BGR->RGB
    # Map class IDs to names
    found = []
    for b in res[0].boxes:
        cls = int(b.cls[0].cpu().numpy())
        name = res[0].names[cls]
        row = drug_db[drug_db["name"].str.lower()==name.lower()]
        if len(row):
            info = row.iloc[0].to_dict()
        else:
            info = {"name": name, "dosage": "N/A", "side_effects": "N/A"}
        info["confidence"] = float(b.conf[0].cpu().numpy())
        found.append(info)
    return ann, pd.DataFrame(found) if found else "No medicines detected."

import gradio as gr
with gr.Blocks() as med_ui:
    gr.Markdown("## üíä MediGuide AI Pharmacist")
    img = gr.Image(type="pil", label="Upload medicine image")
    out_img = gr.Image(type="pil", label="Detections")
    out_tbl = gr.Dataframe(label="Info (from CSV)")
    btn = gr.Button("Detect & Lookup")
    def _run(i):
        det_img, table = med_detect_and_lookup(i)
        if det_img is None:
            return None, pd.DataFrame([{"message": table}])
        return det_img, table if isinstance(table, pd.DataFrame) else pd.DataFrame([{"message": table}])
    btn.click(_run, img, [out_img, out_tbl])
med_ui.launch()
```
